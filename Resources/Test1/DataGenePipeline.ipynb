```python
# training data generation pipeline
import PyPDF2
import openai
import json

with open('10.1.1.1049.6903.pdf', 'rb') as pdf_file:
    reader = PyPDF2.PdfReader(pdf_file)
    text = ''

    for i in range(len(reader.pages)):
        page = reader.pages[i]
        text += page.extract_text()


openai.api_key = ''

def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
    model=model,
    messages=messages,
    temperature=0,
    )
    return response.choices[0].message["content"]

prompt = '''I am fine tuning a LLM. Right now, I have text. And I want you to extract question-answer pair from them. Since I will upload other files. Please insure data quality for fine tuning LLM. Please make sure you found all the question-answer pairs. Please make sure the generated format is in Json format like: [
  {
    "question": ,
    "answer": 
  },
  {
    "question": ,
    "answer": 
  },...'''

qa_pairs = []

tokens_per_chunk = 2048
for i in range(0, len(text), tokens_per_chunk):
    chunk = text[i:i+tokens_per_chunk]
    combined_text = prompt + chunk
    response = get_completion(combined_text)

    try:
        response_json = json.loads(response)
        qa_pairs.extend(response_json)
    except json.JSONDecodeError:
        print(f"Failed to parse response to JSON: {response}")

with open('qa_pairs.json', 'w') as json_file:
    json.dump(qa_pairs, json_file)
```
